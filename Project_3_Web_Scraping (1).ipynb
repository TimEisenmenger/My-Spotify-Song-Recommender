```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import random
```

```python
url = "https://www.billboard.com/charts/hot-100"
```

```python
response = requests.get(url)
response.status_code
```

```python
# Creating soup
soup = BeautifulSoup(response.content, "html.parser")
soup
```

## Creating the BilBoard Top100 DataFrame

```python
# defining 2 empty lists to store content
song_name = []
song_interpret = []
```

```python
# Check length to see how many iterates
iter_count = len(soup.select("span.chart-element__information"))
iter_count
```

```python
# Append input to the lists
for i in range(iter_count):
    song_name.append(soup.select("span.chart-element__information__song")[i].get_text())
    song_interpret.append(soup.select("span.chart-element__information__artist")[i].get_text())
```

```python
# Check if it worked
print(song_name)
```

```python
print(song_interpret)
```

```python
# Creating a dictionary to define column names and what the show
top_100 = pd.DataFrame({"Song Title":song_name,
                       "Song Artist":song_interpret
                      })

# Show DataFrame and check if it worked
top_100.head(15)
```

```python
top_100.shape
```

## Asking for user input and check if this song is in the DataFrame

```python
# Asking for user input
song_title = input("Please tell me a song name! Your choice: ")
```

```python
# Check if user input is in our dataframe
if top_100['Song Title'].str.contains(song_title).any():
    print("Nice! Your song is in the top 100 charts.")
    print("\n")
    yes_or_no = input("Do you want another song request? Please answer with Yes or No. ")
    if yes_or_no == "Yes":
        song_request = random.choice(song_name)
        print("\n")
        print(f"Do you want to hear {song_request} ?")
    else:
        print("Thanks anyway! :)")
        
else:
    print("This song is not in the top 100, sorry!")
```

```python
# Check if input is in top 100 songs:
def song_input(song_title):
    # Both togehter:
    song_title = input("Please tell me a song name! Your choice: ")
    if top_100['Song Title'].str.lower().str.contains(song_title.lower()).any():
        print("\n")
        print("Nice! Your song is in the top 100 charts.")
        print("\n")
        yes_or_no = input("Do you want another song request? Please answer with Yes or No . ")
        # Asking for song request
        if yes_or_no == "Yes":
            song_request = top_100.sample()
            print("\n")
            print("Do you want to hear" ,list (song_request["Song Title"])[0],  "from",list (song_request["Song Artist"])[0],"?")
            print("\n")
            # Playing the new song
            no_or_yes = input("Please answer with Yes or No again. ")
            if no_or_yes == "Yes":
                print("\n")
                print("You are now listening to" ,list (song_request["Song Title"])[0], ". Enjoy!" )
            else:
                print("\n")
                print("Maybe next time.")
        else:
            print("\n")
            print("Thanks anyway! :)")

    else:
        print("This song is not in the top 100, sorry!")
```

```python
# Final function to check if song is in the top 100
song_input(song_title)
```

```python
playlist = sp.user_playlist_tracks("spotify", "1G8IpkZKobrIlXcVPoSIuf")
print(list(playlist.keys()))
len(playlist)
```

```python
def get_playlist_tracks(username, playlist_id):
    results = sp.user_playlist_tracks(username,playlist_id)
    tracks = results['items']
    while results['next']:
        results = sp.next(results)
        tracks.extend(results['items'])
    return tracks
```

```python
playlist = get_playlist_tracks("spotify", "1G8IpkZKobrIlXcVPoSIuf")
```
